**Time Complexity**

- O(n) - Linear time - loops
- O(n*log(n)) - Log Linear - Sorting Operations usually
- O(n^2) - Quadratic - 2 nested loops - compare every element with every other element
- O(2^n) - Exponential - recursion
- O(n!) - adding loop for every element

*What can cause time in a function?*
- Operations (+ - * /)
- Comparisons (> < =)
- Loops
- Function call

**Rule Book**
- Always worst cae
- Drop constants
- Different inputs should have different variables. O(a+b). A and B arrays nested would be O(a*b)
- Drop non-dominat terms

O(n) - Linear Time

```python
def AnotherFunction():
  pass

# 2 + n + n + 1 => 3 + 2n => 2n => n | worst-case, drop constants
#O(n)
def FunChallenge(input):
  a = 10 # O(1)
  a = 50 + 3 # O(1)

  for i in input: # worst-case
    AnotherFunction() # O(n) 
    stranger = True # O(n)
    a = a+1 # O(n)

  for i in input:
    AnotherFunction() # O(n)
  
  return a # O(1)
```

O(1) - Constant Time

```python
# 1+1+1+1+1 => 5 => 1
# O(1)
def PrintFirstFiveItems(input):
    print(input[0]) # O(1)
    print(input[1]) # O(1)
    print(input[2]) # O(1)
    print(input[3]) # O(1)
    print(input[4]) # O(1)
```

O(a+b) - diff terms for inputs - Linear time

```python
#O(a+b)
def FunChallenge(input1, input2):
  a = 10 # O(1)
  a = 50 + 3 # O(1)

  for i in input1: 
    AnotherFunction() # O(a) 
    stranger = True # O(a)
    a = a+1 # O(a)

  for i in input2:
    AnotherFunction() # O(b)
  
  return a # O(1)
```

O(n^2) - Quadratic / O(a*b)

```python
#Log all pairs
x = [1,2,3,4,5]
y = [6,7,8,9,0]

def LogAllPairs(lst1, lst2): #O(a*b)
    for i in lst1: #O(a)
        for y in lst2: #O(b)
            print(i,y)

LogAllPairs(x,y)
```

drop non-dominant
O(n^3) > O(n^2) - Quadratic > O(n) - Linear > O(1) - Constant
